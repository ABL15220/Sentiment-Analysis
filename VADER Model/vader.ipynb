```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('all')

plt.style.use('ggplot')
```

```python
df = pd.read_csv('../data/Reviews.csv')
```

```python
df_og = df.copy()  # original dataframe backup
```

```python
ax = df['Score'].value_counts().sort_index().plot(kind='bar', title='Count of Reviews by Star Rating', figsize=(10, 5))
ax.set_xlabel('Star Rating')
ax.set_ylabel('Number of Reviews')
plt.show()
```
